#KNN MODEL

```python
X = df.drop("churn",1)
Y = df["churn"]
```

```python
corr = df.corr()                   #FINDING THE CORRELATION
```

```python
cor = abs(corr["churn"])
```

```python
features = cor[cor>=0.3]          #SELCETING RELEVENT FEATURES
features
```

```python
x = df[['tenure','employ','callcard','loglong','churn']]                    #selcting features of the model that are helping to predict the accuracy
```

```python
x
```

```python
from sklearn.model_selection import train_test_split                                #importing the file for the train and test set of the dataset
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, Y, test_size=0.2)                #TRAINIG AND TESTING THE DATASET
```

```python
print ('Train set:', X_train.shape, y_train.shape)                     #printing the values of the traing set and test set
print ('Test set:', X_test.shape, y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier              #from sklearn importing the kneighbours classifier 
```

```python
k = 3
n = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
n
```

```python
yhat = n.predict(X_test)                        #PREDICTING THE TEST SET
```

```python
from sklearn import metrics
```

```python
print("Train set Accuracy: ", metrics.accuracy_score(y_train, n.predict(X_train)))                            #CALCULATING THE ACCURACY OF THE MODEL
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat))
```
